```python
#PLOTTING A SPECTOGRAM OF AN AUDIO SAMPLE
stft_audio = librosa.stft(audio)
Xdb = librosa.amplitude_to_db(abs(stft_audio))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')
```
